## 4. Exploratory Data Analysis (EDA)

### Average Call Duration by Education Level

```
df = pd.read_csv("bank-additional-full.csv", sep=';')

# Display the first few rows to confirm correct loading
print(df.head())

# Create a bar chart of average duration by education level
plt.figure(figsize=(10, 6))
df.groupby('education')['duration'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Call Duration by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Duration (seconds)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Distribution of Age

```
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Job Categories Count

```
plt.figure(figsize=(12, 7))
sns.countplot(y='job', data=df, order=df['job'].value_counts().index)
plt.title('Count of Each Job Category')
plt.xlabel('Count')
plt.ylabel('Job Category')
plt.show()
```

### Distribution of Call Duration

```
plt.figure(figsize=(10, 6))
sns.histplot(df['duration'], bins=50, kde=True)
plt.title('Distribution of Call Duration')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.show()
```

### Correlation Heatmap of Numerical Features

```
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## 5. Conclusion

This exploratory data analysis of the bank marketing campaign dataset has provided initial insights into customer demographics, job roles, education levels, and their correlation with call duration and campaign outcomes.

We observed:

*   **Age Distribution:** A typical distribution with a peak in middle age.
*   **Job Categories:** A diverse range of job roles, with some categories having higher counts than others.
*   **Call Duration:** A right-skewed distribution for call duration, indicating many short calls and fewer very long ones. The average call duration varied significantly across different education levels.
*   **Numerical Feature Correlations:** A heatmap revealed relationships between numerical features, which could be important for predictive modeling.

Further analysis could delve into understanding the factors influencing 'y' (subscription to a term deposit) and building predictive models to optimize marketing campaign strategies.

This analysis provides a strong foundation for understanding the dataset and can be extended with more advanced statistical methods and machine learning models to extract deeper insights and actionable strategies.
